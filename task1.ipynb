```python
import pandas as pd
dataset = pd.read_csv('/content/data.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(dataset['Day'],dataset['Tickets_sold'],color='Magenta')
plt.title('Ticket vs price')
```

```python
plt.bar(dataset['Day'],dataset['Benfit'],color='green')
plt.title('Ticket vs price')
```

```python
plt.stackplot(dataset['Day'],dataset['Benfit'],color='yellow')
plt.title('Ticket vs price')
```

```python
plt.bar(dataset['Day'],dataset['Benfit'],color='red',edgecolor='blue')
plt.title('Ticket vs price')
```
