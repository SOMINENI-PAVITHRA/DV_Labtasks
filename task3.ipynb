```python
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.mosaicplot import mosaic
from joypy import joyplot

uploaded = files.upload()
filename = list(uploaded.keys())[0]
df = pd.read_csv('/content/sales_data_multivariate (1).csv')

df['Date'] = pd.to_datetime(df['Date'])

plt.figure(figsize=(12, 6))
df_grouped = df.groupby(['Category', 'Product']).size().unstack()
df_grouped.plot(kind='bar', stacked=True, colormap='viridis', figsize=(12, 6))
plt.title('Stacked Bar Chart: Product vs. Category', fontsize=16)
plt.ylabel('Count', fontsize=12)
plt.xlabel('Category', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.countplot(x='Category', hue='Product', data=df, palette='viridis')
plt.title('Grouped Bar Chart: Product vs. Category', fontsize=16)
plt.ylabel('Count', fontsize=12)
plt.xlabel('Category', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Product')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
mosaic(df, ['Category', 'Product'], title='Mosaic Plot: Product and Category')
plt.show()

plt.figure(figsize=(12, 6))
price_summary = df.groupby('Category')['Price'].agg(['mean', 'median', 'std']).reset_index()
sns.barplot(x='Category', y='mean', data=price_summary, palette='coolwarm', ci=None)
plt.title('Bar Chart: Category vs. Price Mean', fontsize=16)
plt.ylabel('Average Price', fontsize=12)
plt.xlabel('Category', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.kdeplot(data=df, x='Price', hue='Category', multiple='stack', palette='coolwarm')
plt.title('Grouped Kernel Density Plots: Category vs. Price', fontsize=16)
plt.xlabel('Price', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Quantity', data=df, palette='Set2')
plt.title('Box Plot: Category vs. Quantity', fontsize=16)
plt.ylabel('Quantity', fontsize=12)
plt.xlabel('Category', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
sns.violinplot(x='Category', y='Price', data=df, palette='Set3')
plt.title('Violin Plot: Category vs. Price', fontsize=16)
plt.ylabel('Price', fontsize=12)
plt.xlabel('Category', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 8))
joyplot(data=df, by='Category', column='Price', colormap=plt.cm.coolwarm, fade=True)
plt.title('Ridgeline Plot: Category vs. Price', fontsize=16)
plt.xlabel('Price', fontsize=12)
plt.show()

plt.figure(figsize=(12, 6))
sns.swarmplot(x='Category', y='Price', data=df, palette='husl', size=8)
plt.title('Beeswarm Plot: Category vs. Price', fontsize=16)
plt.ylabel('Price', fontsize=12)
plt.xlabel('Category', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Continuous vs. Continuous Analysis

plt.figure(figsize=(12, 6))
sns.scatterplot(x='Price', y='Quantity', data=df, color='b', s=50)
sns.regplot(x='Price', y='Quantity', data=df, scatter=False, color='r', ci=None)
plt.title('Scatterplot with Fit Line: Price vs. Quantity', fontsize=16)
plt.ylabel('Quantity', fontsize=12)
plt.xlabel('Price', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
